## Relational plot

```python
sns.relplot(x="total_bill",y="tip",data=data)
plt.show()
```

## Scatter plot

```python
sns.scatterplot(x="total_bill",y="tip",data=data)
plt.show()
```

## Line plot

```python
sns.lineplot(x="total_bill",y="tip",data=data)
plt.show()
```

```python
sns.lineplot(data=data)
plt.show()
```

## Distribution plot

```python
sns.displot(data["total_bill"])
plt.show()
```

## Histogram plot

```python
sns.histplot(data["total_bill"])
plt.show()
```

## Kernel Density Estimation Plot

```python
sns.kdeplot(data["total_bill"])
plt.show()
```

## Box plot

```python
sns.boxplot(x="total_bill",y="tip",data=data)
plt.show()
```

## Violin plot

```python
sns.violinplot(x="total_bill",y="tip",data=data)
plt.show()
```

## Bar plot

```python
sns.barplot(x="total_bill",y="tip",data=data)
plt.show()
```

## Count plot

```python
sns.countplot(x="total_bill",data=data)
plt.show()
```
