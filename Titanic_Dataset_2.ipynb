## Relational plot

```python
sns.relplot(x="age",y="class",data=data)
plt.show()
```

## Scatter plot

```python
sns.scatterplot(x="age",y="class",data=data)
plt.show()
```

## Line plot

```python
sns.lineplot(x="age",y="class",data=data)
plt.show()
```

```python
sns.lineplot(data=data)
plt.show()
```

## Distribution plot

```python
sns.displot(data["age"])
plt.show()
```

## Histogram plot

```python
sns.histplot(data["age"])
plt.show()
```

## Kernel Density Estimation Plot

```python
sns.kdeplot(data["age"])
plt.show()
```

## Box plot

```python
sns.boxplot(x="age",y="class",data=data)
plt.show()
```

## Violin plot

```python
sns.violinplot(x="age",y="class",data=data)
plt.show()
```

## Bar plot

```python
sns.barplot(x="age",y="class",data=data)
plt.show()
```

## Count plot

```python
sns.countplot(x="age",data=data)
plt.show()
```

## Heatmap plot

```python
corr=data.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
```

## pair plot

```python
sns.pairplot(data)
plt.show()
```
