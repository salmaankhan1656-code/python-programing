**List Items**

List items are ordered, changeable, and allow duplicate values.

List items are indexed, the first item has index [0], the second item has index [1] etc.

```python
list1 = ["a", "b", "c", "a", "c"]
print(list1)
```

```python
#Access List items
list1[1]
```

```python
list1[2]
```

```python
#Range of Indexes
list1[2:4]
```

```python
#Negative Indexing
list11 = ["a", "b", "c", "d", "e"]
list11[-2]
```

```python
print(list11[-4:-1])
```

```python
list11[-5:]
```

```python
list11[-2:]
```

```python
# Change item value
L2=[1,2,3,4,5,6,7]
print(L2)
```

```python
L2[2]=8
print(L2)
```

```python
L2[1:3]="A","B"
print(L2)
```

```python
L2[2]=2
print(L2)
```

```python
L3=[6,7,8,9]
L3[2]=13
print(L3)
```

```python
#List item Addition
# Append Items
# To add an item to the end of the list, use the append() method:
```

```python
# List creation
```

```python
course=["Python","SQL","Stats"]
print(course)
```

```python
course.append("Excel")
print(course)
```

```python
# Insert Items
# To insert a list item at a specified index, use the insert() method
```

```python
course.insert(3,'Algo')
print(course)
```

```python
company=['TCS','Infosys','Wipro']
print(company)
```

```python
# Extend List
# To append elements from another list to the current list, use the extend() method.
```

```python
course.extend(company)
```

```python
print(course)
```

```python
# Remove Specified Item
# The remove() method removes the specified item.
```

```python
course.remove('Algo')
print(course)
```

```python
#Pop Method
#If do not specify the index, the pop() method removes the last item.
stud=['M','N','O','P','Q','R']
stud.pop()
```

```python
stud
```

```python
#pop() method removes the specified index.
stud.pop(1)
```

```python
stud
```
