 ***if Statement***

```python
# For True case
age = 20
if age >= 18:
    print("You are eligible to vote.")
```

```python
# For False case
age = 12
if age >= 18:
    print("You are eligible to vote.")
```

```python
# Not following indentation
age = 20
if age >= 18:
print("You are eligible to vote.")
```

**if-else Statement**

```python
score = 75
if score >= 60:
    print("You passed the exam.")
else:
    print("You did not pass the exam.")
```

**if-elif-else Statement**

```python
#To check grade of Marks.
grade = 85
if grade >= 90:
    print("Grade: A")
elif grade >= 80:
    print("Grade: B")
elif grade >= 70:
    print("Grade: C")
else:
    print("Grade: F")
```

**[Important Rules]**

**Indentation:** Python uses indentation to define code blocks within conditional statements. Consistent indentation (typically four spaces) is crucial for correct execution.

**Colon:** Each if, elif, and else line must end with a colon (:).
Logical Operators: You can combine multiple conditions using logical operators like and, or, and not within conditional statements.



```python
#To Check whether given number is positive or not?
number = int(input('Enter a number: '))

# check if number is greater than 0
if number > 0:
    print(f'{number} is a positive number.')

```

```python
#Greater than between Two numbers
a = 200
b = 33
if b > a:
  print("b is greater than a")
else:
  print("b is not greater than a")
```

```python
# Python program to check if the input number is odd or even.
# A number is even if division by 2 gives a remainder of 0.
# If the remainder is 1, it is an odd number.

num = int(input("Enter a number: "))
if (num % 2) == 0:
   print("Number is even")
else:
   print("Number is odd")
```
